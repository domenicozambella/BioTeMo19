# Multiple testing problem (un esempio)

Abbiamo una moneta con $2$ facce $H$ e $T$. Vogliamo fare i seguenti test di ipotesi per $i=H,T$:

$H_{0,i}:$ la probabilità che esca $i$ è $=1/2$

$H_{A,i}:$ la probabilità che esca $i$ è $>1/2$

<br>
## Caso A

Eseguiamo due volte un test binomiale. Assumiamo che la moneta sia equilibrata. Qual'è la probabilità di rigettare erroneamente almeno una delle due ipotesi nulle con una significatività del $5\%$?

<br>
## Risposta A

È come ripetere un esperimento Bernoulliano 2 volte con probabilità di successo $5\%$. La probabilità di non avere alcun successo è $0.95^{2}= 90.25\%$. Quindi la probabilità di rigettare almeno un'ipotesi nulla è $\cong 9.75\%$

<br>
## Caso B

Eseguiamo due test binomiali ma usando la stessa sequenza di lanci. Assumiamo che la moneta sia equilibrata. Qual'è la probabilità di rigettare erroneamente almeno una delle due ipotesi nulle con una significatività del $5\%$?

<br>
## Risposta B

I due test non possono entrambi risultare positivi. La probabilità che almeno uno dei due test sia positivo è la somma della probabilità che uno sia positivo e quindi la probabilità che almeno uno sia positivo è $5\%+5\%\ =\ 10\%$?

<br>
## Osservazione

Se abbiamo più di due test la differenza tra caso A e del caso B è ben più estrema. 

Rivediamo in modo più generale i due casi considerati sopra.

**A.**   Facciamo $m$ test **indipendenti**.  Assumiamo che tutte le ipotesi nulle siano vere. Sia $\alpha$ è l'errore del I tipo nel singolo test. La probabilità di rigettare almeno almeno un'ipotesi nulla è

$\qquad\alpha_{A,m}\ =\ 1 - (1-\alpha)^m$


**B.**   Facciamo $m$ test mutualmente **esclusivi**.  Assumiamo che tutte le ipotesi nulle siano vere. Sia $\alpha$ è l'errore del I tipo nel singolo test. La probabilità di rigettare almeno almeno un'ipotesi nulla è

$\qquad\alpha_{B,m}\ =\ m\alpha$

Per un esempio numerico poniamo $\alpha=0.05$ ed $m=10$, otteniamo

$\qquad\alpha_{A,m}\ =\ 40\%$

$\qquad\alpha_{B,m}\ =\ 50\%$

I due scenari danno risultati diversi ma in entrambi i casi la probabilità di avere (almeno) un falso positivo è ben superiore al $5\%$.
